```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/advertising (1).csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
print(df.shape)
```

```python
#View descriptive statistics
print(df.describe())
```

```python
#Outlier Analysis

fig, axs = plt.subplots(3, figsize = (5,5))
plt.subplots_adjust(hspace = .5)
sns.boxplot(df['TV'], ax = axs[0])
sns.boxplot(df['Radio'], ax = axs[1])
sns.boxplot(df['Newspaper'], ax = axs[2])
plt.tight_layout()
```

```python
#declare feature variables and target variables
X=df['TV']
y=df['Sales']
```

```python
#plot scatter plot between X and Y
plt.scatter(X,y,color='blue',label='Scatter Plot')
plt.title('Relationship between TV and Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
#Print Dimensions of X and Y
print(X.shape)
print(y.shape)
```

```python
# changing the dimension to array
X=np.array(X)
y=np.array(y)
```

```python
#Reshape X and y
X=X.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
#print the dimension of X and Y after reshaping
print(X.shape)
print(y.shape)
```

```python
#splitting the data
#random_state-- the set of data does not change from sklearn.model_selection import train
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
y_pred=lm.predict(X_test)
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,lm.predict(X_train),color='blue')
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,lm.predict(X_train),color='blue')
plt.title('Test set result')
plt.xlabel('TV')
plt.ylabel('sales')
plt.show()
```

```python
slope = lm.coef_
intercept = lm.intercept_,
print("estimate model slope:",slope)
print("estimate model intercept:",intercept)
```
